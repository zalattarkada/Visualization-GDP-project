```python
import requests
import pandas as pd

url = "http://api.worldbank.org/v2/country/all/indicator/SP.POP.TOTL?format=json&per_page=20000"
response = requests.get(url)

if response.status_code == 200:
    data = response.json()[1]  # Get the data from the second element of the response
    df = pd.json_normalize(data)  # Convert the JSON data into a pandas DataFrame
    print(df.head())
else:
    print("Request failed with status code:", response.status_code)
```

```python
df = df.rename(columns={'countryiso3code': 'country_code', 'indicator.id': 'indicator_id', 'indicator.value': 'indicator_value', 'date': 'year', 'country.value': 'Country', 'value':'population'})
```

```python
df.to_csv('../data/population.csv')
```
